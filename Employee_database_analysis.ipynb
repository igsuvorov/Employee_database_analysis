## Проверим гипотезу, что сотрудники с высоким окладом проводят на работе больше времени

```python
df.head()
```

Сначала проверим с помощью теста Стьюдента для двух выборок

```python
salary_high = df[df['salary'] == 'high']
salary_high.head()
```

```python
salary_high = salary_high.loc[:,['average_montly_hours']]
salary_high.head()
```

```python
other_salary = df[(df['salary'] == 'medium') | (df['salary'] == 'low')]
other_salary.head()
```

```python
other_salary = other_salary.loc[:,['average_montly_hours']]
other_salary.head()
```

Н0 - обе выборки имеют равные средние(сотрудники с высоким окладом не проводят больше времени на работе)

```python
from scipy import stats 

stat, p = stats.ttest_ind(salary_high, other_salary)

print('Статистика =', stat, 'p =', p)

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, средние, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, средние, вероятно, различаются')
```

Сравним средние

```python
salary_high.mean()
```

```python
other_salary.mean()
```

```python
# посмотрим визуально
sns.boxplot(y = 'average_montly_hours', data = salary_high);
sns.boxplot(y = 'average_montly_hours', data = other_salary);
```

Вывод: проведя статистический тест, мы можем придти к выводу, что сотрудники с высоким окладом проводят столько  же времени на работе(или даже меньше), как и другие сотрудники



Проверим гипотезу с помощью дисперсионного анализа

```python
fvalue, pvalue = stats.f_oneway((df['salary']== 'high'),(df['salary']== 'medium'), (df['salary']== 'low'))

print('Статистика=%.3f, p=%.6f' % (fvalue, pvalue))
if p > 0.05:
    print('Не отклоняем нулевую гипотезу, средние, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, средние, вероятно, различаются')
```

```python
sns.boxplot(x='salary', y='average_montly_hours', data=df, palette='Set1')
```

Вывод, дисперсионный анализ подтверждает предыдущие выводы

## Разделим данные на тестовую и обучающую выборки. Построим модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и иsalary).Оценим качество модели на тестовой выборке

```python
df.head()
```

```python
# оставим все метрики без department и salary
df_new = df[df.columns[0:8]]
df_new.head()
```

```python
# будем прогнозировать на основе уволившихся и сразу передадим в массив
quit = pd.array(df_new['left'])
quit
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_new, quit, test_size = 0.30)
```

```python
len(y_test)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
```

```python
# обучаем данные
lda.fit(X_train, y_train)
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
# смотрим разницу факта и прогноза
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
```

```python
result
```

```python
# расчет точности - отношение верных прогнозов к общему количеству позиций
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lda.predict(X_test))
```

Вывод:судя по тому, что accuracy_score = 0.76, мы не можем утверждать, что прогнозная модель имеет высокое качество
